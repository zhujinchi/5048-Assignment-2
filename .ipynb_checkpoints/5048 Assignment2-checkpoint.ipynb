## Data preprocessing
I preprocessed the data before using it, removing rows with empty data and rows with corrected data

```python
import csv

with open('new_clean.csv') as f:
    f_csv = csv.reader(f)
    
    #get headers
    headers = next(f_csv)
    
    total_data = []
    
    #loop for content
    for row in f_csv:
        if row[1] != 'None' and row[2] != '' and row[3] != '' and row[3] != 'Correctional settings'  and row[4] != '' and row[5] != '':
            total_data.append(row)
```

```python
#get total lines
count = len(total_data)
print("total lines:"+str(count))

total_postcode = []
total_lhd_2010_code = []
total_lhd_2010_name = []
total_lga_code19 = []
total_lga_name19 = []
#get data for each partition
for i in range(count):
    total_postcode.append(total_data[i][1])
    total_lhd_2010_code.append(total_data[i][2])
    total_lhd_2010_name.append(total_data[i][3])
    total_lga_code19.append(total_data[i][4])
    total_lga_name19.append(total_data[i][5])

total_postcode = set(total_postcode)
total_lhd_2010_code = set(total_lhd_2010_code)
total_lhd_2010_name = set(total_lhd_2010_name)
total_lga_code19 = set(total_lga_code19)
total_lga_name19 = set(total_lga_name19)

#postcode
print("\n==>number of total_postcode: "+str(len(total_postcode)))
print(total_postcode)

#lhd_2010_code
print("\n==>number of total_lhd_2010_code: "+str(len(total_lhd_2010_code)))
print(total_lhd_2010_code)

#lhd_2010_name
print("\n==>number of total_lhd_2010_name: "+str(len(total_lhd_2010_name)))
print(total_lhd_2010_name)

#lga_code19
print("\n==>number of total_lga_code19: "+str(len(total_lga_code19)))
print(total_lga_code19)

#lga_name19
print("\n==>number of total_lga_name19: "+str(len(total_lga_name19)))
print(total_lga_name19)
```

# Visualization

## Chart data pre-processing

```python
x_axis = []
y_axis_list = []

for i in range(len(total_data)):
    if total_data[i][0] not in x_axis:
        x_axis.append(total_data[i][0])

for d in range(len(lga_name)):
    y_axis = []
    data = total_data_sortBy_lga_name[lga_name[d]]
    
    for i in range(len(x_axis)):
        y_axis.append(0)
    
    for i in range(len(data)):
        if data[i][0] in x_axis:
            for j in range(len(x_axis)):
                if data[i][0] == x_axis[j]:
                    y_axis[j] += 1
   
    y_axis_list.append(y_axis)
    
x_axis_gs = []
y_axis_list_gs = []

for i in range(len(total_data)):
    if total_data[i][0] not in x_axis_gs:
        x_axis_gs.append(total_data[i][0])

for d in range(len(lga_name_new)):
    y_axis = []
    data = total_data_sortBy_lga_name_new[lga_name_new[d]]
    
    for i in range(len(x_axis)):
        y_axis.append(0)
    
    for i in range(len(data)):
        if data[i][0] in x_axis:
            for j in range(len(x_axis)):
                if data[i][0] == x_axis[j]:
                    y_axis[j] += 1
   
    y_axis_list_gs.append(y_axis)
```

## Build the chart

```python
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.globals import ThemeType
import pandas as pd
from pyecharts.charts import Bar,Grid
from pyecharts import options as opts

line=(
    Line(init_opts=opts.InitOpts(theme=ThemeType.MACARONS))
    .add_xaxis(xaxis_data = x_axis_gs)
    .set_global_opts(title_opts = opts.TitleOpts(title = 'NSW Outbreak infection case trends')))

for i in range(len(lga_name_new)):
    line.add_yaxis(str(lga_name_new[i]), y_axis_list_gs[i], label_opts = opts.LabelOpts(is_show = False), is_smooth = False)
line.render_notebook()
```

```python
stack_bar = (
    Bar(init_opts=opts.InitOpts(theme=ThemeType.MACARONS))
    .add_xaxis(x_axis)
    .add_yaxis(lga_name_new[0], y_axis_list_gs[0],bar_min_width=1,bar_max_width=1,color="#a834a8", stack="stack")
    .add_yaxis(lga_name_new[1], y_axis_list_gs[1],color="#42b7bd", stack="stack",bar_min_width=1,is_selected=True)
    #
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False))     
    .set_global_opts(title_opts = opts.TitleOpts(title = 'NSW Outbreak infection case trends'))
)

stack_bar.render_notebook()
```

Change the time unit of the data to a 7-day interval (the number of infections within 7 days is pooled into a single day display). The time span is from March 1, 2020 to October 24, 2021

```python
import time,datetime
from datetime import date
    
x_axis_week = [] 

def getday(the_day): 
      result_date = the_day + datetime.timedelta(days = 7)
      d = result_date.strftime('%Y-%m-%d')
      return d
   
day_start = '2020-03-01'
day_end = datetime.datetime.strptime('2021-10-17', '%Y-%m-%d').date()
while True:
      x_axis_week.append(day_start)
      the_date = datetime.datetime.strptime(day_start, '%Y-%m-%d').date()
      day_start = getday(the_date)
      if (the_date-day_end).days > 0:
          break
```

```python
y_axis_list_new = []

x_axis_copy = x_axis

j = 0
i = 0
while True:
    time1 = datetime.datetime.strptime(x_axis_week[j], '%Y-%m-%d').date()
    time2 = datetime.datetime.strptime(x_axis_copy[i], '%Y-%m-%d').date()
    
    if (time2-time1).days <= 6:
        x_axis_copy[i] = x_axis_week[j]
        if i < 545:
            i = i+1
    else:
        j = j+1
        if j == 84:
            break
```

Timeline and processed weekly data showing the trend of the epidemic in NSW, analysis of the extent of the epidemic in each region, and the impact of government policies on the trend, see report for details

Due to browser performance, I could not present all 1442 data in one timeline, so I split it into two maps to show

```python
import random
import math

#change color with the cases
def RGB_to_Hex(rgb):
    RGB = rgb.split(',')    
    color = '#'
    for i in RGB:
        num = int(i)
        color += str(hex(num))[-2:].replace('x', '0').upper()
    return color

def Get_color(n):
    n = 100+155*(n/974)
    n = 255-n
    m = math.ceil(n)
    return RGB_to_Hex('255,'+str(m)+','+str(m)) 
```

```python
import folium
from folium.plugins import TimestampedGeoJson

NSW_map_1 = folium.Map(location=[-33.2532, 146.9211], 
    zoom_start=6)
NSW_map_2 = folium.Map(location=[-33.2532, 146.9211], 
    zoom_start=6)

polygon_array_1 = []
polygon_array_2 = []

#range from 0-1442
for i in range(1442):
    polygon = {
    'type': 'Feature',
    'geometry': {
        'type': 'MultiPolygon',
        'coordinates': [usw_cities['features'][maps_array[i]['coordinates']]['geometry']['coordinates']],
    },
    'properties': {
        'style': {
            'color': 'red',
            'weight': 0.1
        },
        'times': [ maps_array[i]['dates'], maps_array[i]['dates']+'T23:59:59']
        }
    }
    if i <= 721:
        polygon_array_1.append(polygon)
    else:
        polygon_array_2.append(polygon)

    
TimestampedGeoJson(
    {'type': 'FeatureCollection', 'features': polygon_array_1},
    period='P1W',
    duration='P1W',
    auto_play=False,
    loop=False,
    loop_button=True,
    date_options='YYYY/MM/DD',
).add_to(NSW_map_1)

TimestampedGeoJson(
    {'type': 'FeatureCollection', 'features': polygon_array_2},
    period='P1W',
    duration='P1W',
    auto_play=False,
    loop=False,
    loop_button=True,
    date_options='YYYY/MM/DD',
).add_to(NSW_map_2)
```
